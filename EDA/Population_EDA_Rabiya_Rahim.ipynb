### Visualizing the missing values

Normally we can't drop the missing values.Instead dropping we need to fill the missing value with anyother suitable value

```python
sns.heatmap(data.isnull(),cbar=False,cmap='viridis')
```

```python
data.fillna(method='bfill',inplace=True)
```

```python
data.isnull().sum()
```

### Visualization

```python
data.plot()
```

### Average of Annual Change

```python
data["Annual_Change"].mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['Annual_Change'], color='blue', marker='o', markersize=4)
plt.title('Annual Change in Population of Algieria (1950 - 2023)')
plt.xlabel('Year')
plt.ylabel('Annual Change (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Define the bin edges
bin_edges = [2, 3, 4, 5, 6, 7, 8, 9, 10] 

# Create the histogram
plt.figure(figsize=(12, 6))
sns.histplot(data['Annual_Change'], kde=True, bins=[2, 3, 4, 5, 6, 7, 8, 9, 10], color='blue', edgecolor='black')

# Adding labels and title
plt.xlabel('Annual Change (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Annual Change in Population of Algeria')

# Show the plot
plt.show()
```
